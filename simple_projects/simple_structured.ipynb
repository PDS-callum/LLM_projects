```python
from langchain_ollama import ChatOllama

llm = ChatOllama(
    model="llama3.2",  # any model you have in Ollama (`ollama pull <model>`)
    base_url="http://localhost:11434",  # default for `ollama serve`
)
```

```python
from pydantic import BaseModel, Field

class searchQuery(BaseModel):
    search_query: str = Field(description="An optimised query for someone to enter into Google.")
    justification: str = Field(description="A one-sentence justification for that query.")

structured_llm = llm.with_structured_output(searchQuery)

output = structured_llm.invoke(
    "What is the best way to book a train ticket? Return valid JSON only."
)
print(output.search_query)
print(output.justification)
```
