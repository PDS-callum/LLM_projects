```python
from langchain_ollama import ChatOllama

llm = ChatOllama(
    model="llama3.2",  # any model you have in Ollama (`ollama pull <model>`)
    base_url="http://localhost:11434",  # default for `ollama serve`
)
```

```python
def multiply(a: int, b: int) -> int:
    return a*b
def addition(c: int, d: int) -> int:
    return c+d

# Force the model to call a tool instead of answering directly
llm_w_tool = llm.bind_tools([multiply,addition], tool_choice="required")

output = llm_w_tool.invoke("What is 2 add 3?")
print(output.tool_calls)
```
