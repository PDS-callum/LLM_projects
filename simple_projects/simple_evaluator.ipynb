```python
from pydantic import BaseModel, Field
from typing_extensions import Literal
from langchain_core.messages import HumanMessage, SystemMessage

class feedback(BaseModel):
    claim: str = Field(
        description="The user's claim, reproduced exactly."
    )
    summary: str = Field(
        description="A brief summary of what the claim asserts."
    )
    factuality: Literal["Fact", "Lie"] = Field(
        description=(
            "Fact = the claim is FACTUALLY TRUE (correct). "
            "Lie = the claim is FACTUALLY FALSE (incorrect). "
            "Base your answer on real-world facts only."
        )
    )

fact_detector = llm.with_structured_output(feedback)

EVAL_PROMPT = (
    "Evaluate whether the following claim is factually TRUE or FALSE. "
    "Use Fact only if the claim is correct; otherwise use Lie.\n\nClaim: {claim}"
)

result = fact_detector.invoke([
    SystemMessage(content="You are a fact-checker. Judge only factual accuracy."),
    HumanMessage(content=EVAL_PROMPT.format(claim="The Eiffel Tower is in Germany.")),
])
result
```
